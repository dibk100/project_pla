topikGrade 기준으로 샘플 추출

```python
import os
import json
from IPython.display import Audio, display

meta_root = "meta_data/1. 한국일반"  
audio_root = "raw_data/1. 한국일반"

samples_by_topik = {}

# 추출하고 싶은 TOPIK 등급 리스트 (필요에 따라 수정 가능)
target_topik_grades = ["1", "2", "3", "4", "5", "6"]

for filename in os.listdir(meta_root):
    if not filename.endswith(".json"):
        continue
    filepath = os.path.join(meta_root, filename)
    with open(filepath, "r", encoding="utf-8") as f:
        data = json.load(f)
    
    topik_grade = data.get("skill_info", {}).get("topikGrade")
    if topik_grade in target_topik_grades and topik_grade not in samples_by_topik:
        samples_by_topik[topik_grade] = data
    
    # 다 찾았으면 종료
    if len(samples_by_topik) == len(target_topik_grades):
        break

# 샘플 출력 및 재생
for grade in target_topik_grades:
    sample = samples_by_topik.get(grade)
    if sample is None:
        print(f"TOPIK 등급 '{grade}' 샘플을 찾지 못했습니다.")
        continue
    
    wav_file = sample.get("fileName")
    wav_path = os.path.join(audio_root, wav_file)
    
    print(f"\n==== TOPIK 등급: {grade} ====")
    print(f"파일명: {wav_file}")
    print(f"화자 ID: {sample.get('SpeakerID')}")
    print(f"녹음 정보: {sample.get('file_info')}")
    print(f"발화 - Reading: {sample['transcription'].get('Reading')}")
    print(f"발화 - ReadingLabelText: {sample['transcription'].get('ReadingLabelText')}")
    print(f"발화 - Question: {sample['transcription'].get('Question')}")
    print(f"발화 - AnswerLabelText: {sample['transcription'].get('AnswerLabelText')}")
    print(f"기본 정보: {sample.get('basic_info')}")
    print(f"거주 정보: {sample.get('residence_info')}")
    print(f"스킬 정보: {sample.get('skill_info')}")
    
    if os.path.exists(wav_path):
        display(Audio(wav_path))
    else:
        print("오디오 파일을 찾을 수 없습니다.")
```
